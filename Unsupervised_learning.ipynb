```python
import pandas as pd
import numpy as np
```

```python
network_ads=pd.read_excel('/content/sample_data/SocialNetworkAds.csv.xlsx')
network_ads
```

iloc use to select certain values , here we are assigning which values should x and y should have

```python
x= network_ads.iloc[:,[2,3]].values
y= network_ads.iloc[:,4].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 1/3)
```

here we are cleaning data , standard deviation if scattered then properly arranging for training

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knclassifier = KNeighborsClassifier(n_neighbors=5)
knclassifier.fit(x_train , y_train)
```

predicting on the basis of x i.e independent variable

```python
y_predict= knclassifier.predict(x_test)
```

k value changes accuracy changes

```python
from sklearn import metrics
print("accuracy:", metrics.accuracy_score(y_test,y_predict))
```
